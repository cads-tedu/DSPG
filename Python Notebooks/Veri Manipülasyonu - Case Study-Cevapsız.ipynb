<b>18)</b> Ülkelerin kazandığı madalya sayıları ile gayri safi yurtiçi hasılaları arasında bir ilişki olabilir mi?
    
<b>İpucu:</b>
1. Olimpiyatlar kolonunda ülkeler saydırılır..value_counts()

5. En son elde edilen Dataframe, kunye veri setine right join ile birleştirilir. .merge(kunye, how = 'right')
6. En son elde edilen Dataframe, 'Madalya' değişkenine göre büyükten küçüğe sıralanır. .sort_values('Madalya', ascending = False)
</div>

Beklediğimiz gibi, başarılı ülkelerin, gayri safi yurtiçi hasıla miktarları da oldukça yüksek ülkelerdir. Bu, kazanılan madalya sayısına dolaylı olarak etki edebilir. Her konuda daha fazla imkan sağlayabilme vs.

<div class="alert alert-block alert-warning">
<b>19)</b> Ülkelerin kazandıkları madalyaların ortanca değerinden fazla madalya kazanan ülkelerin ortalama, ortanca, minimum ve maximum gayri safi yurtiçi hasıla miktarları nedir?
    
<b>İpucu:</b>
1. Bir önceki soruda oluşturulan Dataframe'e gdp ismi verelim.
2. gdp veri setini madalya sayıları medyan madalya sayısından büyükleri kapsayacak şekilde filtreleyelim. gdp.Madalya.median()
3. .agg() metodu kullanarak ilgili istatistikleri elde edelim.
</div>

```python
#Buraya Kodlayın
```

```python
gdp = pd.DataFrame(olimpiyatlar['Country'].value_counts()).reset_index().rename(columns = {'index':'Country', 'Country':'Madalya'}).merge(kunye, how = 'right').sort_values('Madalya',ascending = False)

gdp[gdp['Madalya'] > gdp.Madalya.median()]['GDP'].agg({'Maks_GDP':'max',
                                                      'Min_GDP':'min',
                                                      'Medyan_GDP':'median',
                                                      'Ortalama_GDP':'mean'})

#veya

gdp[gdp['Madalya'] > gdp.Madalya.median()].agg(Maks_GDP = ('GDP', max),
                                              Min_GDP = ('GDP', min),
                                              Medyan_GDP = ('GDP', np.median),
                                              Ortalama_GDP = ('GDP', np.mean))
```

<div class="alert alert-block alert-warning">
<b>20)</b> Ülkemiz hangi şehirde ve hangi spor dallarında madalya kazanmıştır?
    
<b>İpucu:</b>
1. Olimpiyatlar veri seti ülke olarak sadece Türkiye olacak şekilde filtrelenir.
2. Filtrelenen veri seti 'City' ve 'Sport' kolonlarına göre gruplanır. .groupby()
3. .agg() fonksiyonu kullanarak Madalya_Sayısı adında bir kolon oluşturulur 'Medal' kolonuna 'count' fonksiyonu Python'da varsayılan olarak bulunmadığı için fonksiyon lambda fonksiyonuyla oluşturulup bu kolona uygulanır. lambda x: x.count()
</div>

```python
#Buraya Kodlayın
```

```python
olimpiyatlar[olimpiyatlar['Country'] == 'Turkey'].groupby(['City', 'Sport']).agg(Madalya_Sayısı = ('Medal', lambda x: x.count()))
#Bonus
olimpiyatlar[olimpiyatlar['Country'] == 'Turkey'].pivot_table(values = 'Medal', index='City', columns='Sport', aggfunc=lambda x: x.count())
```
